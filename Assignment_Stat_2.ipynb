Q3. What is the difference between discrete and continuous distributions?

- Discrete Random Variables
  - A discrete random variable takes on a countable number of distinct values.

  - The possible values can be finite or countably infinite, but each value is distinct.

  - Probability Distribution: Discrete random variables are typically described by a probability mass function (PMF), where each value has a specific probability.
- Continuous Random Variables
  - A continuous random variable can take on an infinite number of values within a given range or interval.

  - These values are typically real numbers and can represent measurements.

  - Probability Distribution: Continuous random variables are described by a probability density function (PDF). The probability that the variable takes an exact value is always zero, and probabilities are calculated over intervals (e.g., between 2.0 and 3.0).

Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem (CLT) is a fundamental concept in statistics that describes how the distribution of sample means (or sums) behaves when we repeatedly take random samples from a population.
- Why the Central Limit Theorem is Critical in Statistics:
  - Makes Inference Possible: The CLT enables statistical inference, which is the process of making predictions or generalizations about a population based on sample data. Thanks to the CLT, we can apply the properties of the normal distribution (such as using Z-scores) even when the population data isn't normally distributed, as long as the sample size is sufficiently large.

  - Normality Assumption: Many statistical tests, like t-tests or confidence intervals, rely on the assumption that the sample mean follows a normal distribution. The CLT ensures that, for large enough sample sizes, this assumption holds true, regardless of the original population's distribution.

```python
#Q6.  Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
a = 1  # Lower bound
b = 10  # Upper bound

# Values for the discrete random variable
x = np.arange(a, b+1)

# Probability mass function (PMF) for the discrete uniform distribution
pmf = np.ones_like(x) / len(x)

# Cumulative distribution function (CDF)
cdf = np.cumsum(pmf)

# Plotting the CDF
plt.step(x, cdf, where='post', label="CDF", color='blue', linestyle='-', linewidth=2)

# Adding labels and title
plt.title('Cumulative Distribution Function (CDF) of a Discrete Uniform Distribution')
plt.xlabel('x')
plt.ylabel('CDF(x)')
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.7)


plt.show()
```

```python
#Q9. Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    # Calculate the mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate the Z-scores
    z_scores = (data - mean) / std_dev

    return z_scores

# Example dataset (can be any numeric data)
data = np.random.normal(loc=0, scale=1, size=1000)  # Generating random data

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.plot(z_scores, color='blue', alpha=0.7)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)  # Line at Z=0 for reference
plt.title('Z-scores of the Dataset')
plt.xlabel('Index')
plt.ylabel('Z-score')
plt.show()
```

```python
#Q18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to calculate the Z-score for a given data point
def calculate_z_score(x, mean, std_dev):
    return (x - mean) / std_dev

# Example dataset (can be any numeric data)
data = np.random.normal(loc=0, scale=1, size=1000)  # Generating random data

# Given data point
data_point = 1.5

# Calculate mean and standard deviation of the dataset
mean = np.mean(data)
std_dev = np.std(data)

# Calculate the Z-score for the given data point
z_score = calculate_z_score(data_point, mean, std_dev)

# Calculate the cumulative probability (CDF) for the Z-score from a standard normal distribution
cdf_value = stats.norm.cdf(z_score)

# Print the Z-score and the corresponding cumulative probability (CDF)
print(f"Z-score for the data point {data_point}: {z_score}")
print(f"Cumulative probability for Z = {z_score}: {cdf_value}")

# Plot the standard normal distribution and mark the Z-score on the plot
x_values = np.linspace(-4, 4, 1000)  # X-axis values for plotting
y_values = stats.norm.pdf(x_values)  # Y-axis values for the standard normal distribution

plt.plot(x_values, y_values, label='Standard Normal Distribution', color='blue')
plt.axvline(x=z_score, color='red', linestyle='--', label=f'Z-score: {z_score:.2f}')
plt.title('Standard Normal Distribution and Z-score')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q19. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import scipy.stats as stats

# Function to perform hypothesis testing using Z-statistics
def hypothesis_testing_z(sample, population_mean, population_std, alpha=0.05):
    # Calculate the sample mean and sample size
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value from the Z-statistic
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed test

    # Print the results
    print(f"Sample Mean: {sample_mean}")
    print(f"Z-statistic: {z_statistic}")
    print(f"P-value: {p_value}")

    # Determine whether to reject the null hypothesis
    if p_value < alpha:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

# Example data
np.random.seed(0)
sample = np.random.normal(loc=100, scale=15, size=50)  # Sample from a normal distribution
population_mean = 105  # Population mean (hypothesized)
population_std = 15   # Population standard deviation

# Perform hypothesis testing using Z-statistics
hypothesis_testing_z(sample, population_mean, population_std, alpha=0.05)
```

```python
#Q20. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

# Sample dataset
data = np.array([12, 15, 14, 10, 13, 16, 19, 21, 18, 17])

# Parameters
confidence_level = 0.95  # Confidence level (95%)
n = len(data)  # Sample size
mean = np.mean(data)  # Sample mean
std_dev = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for sample std deviation)

# Degrees of freedom
df = n - 1

# Calculate the t-value for the desired confidence level
t_value = stats.t.ppf((1 + confidence_level) / 2, df)

# Calculate the margin of error
margin_of_error = t_value * (std_dev / np.sqrt(n))

# Confidence interval
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

# Output the result
print(f"Sample Mean: {mean:.2f}")
print(f"Sample Standard Deviation: {std_dev:.2f}")
print(f"Confidence Interval (95%): {confidence_interval[0]:.2f} to {confidence_interval[1]:.2f}")
```

```python
#Q21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
import scipy.stats as stats

# Parameters for the normal distribution
mu = 50  # Mean of the distribution
sigma = 10  # Standard deviation of the distribution
n = 100  # Sample size
confidence_level = 0.95  # Confidence level (95%)

# Step 1: Generate random data from a normal distribution
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate the sample mean and sample standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation (use ddof=1 for unbiased estimator)
standard_error = sample_std / np.sqrt(n)

# Step 3: Calculate the z-score for the given confidence level
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)  # For two-tailed confidence interval

# Step 4: Calculate the margin of error
margin_of_error = z_score * standard_error

# Step 5: Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Display the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Standard Error: {standard_error:.2f}")
print(f"Confidence Interval ({confidence_level*100}%): ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
```

```python
#Q22. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for the normal distribution
mean = 0          # Mean of the distribution
std_dev = 1       # Standard deviation of the distribution

# Generate a range of values (e.g., between -5 and 5)
x_values = np.linspace(-5, 5, 1000)

# Calculate the PDF of the normal distribution for these values
pdf_values = stats.norm.pdf(x_values, loc=mean, scale=std_dev)

# Plot the PDF
plt.plot(x_values, pdf_values, label=f'Normal Distribution\nMean={mean}, Std={std_dev}', color='blue')
plt.fill_between(x_values, pdf_values, color='skyblue', alpha=0.4)

# Add labels and title
plt.title('Probability Density Function (PDF) of a Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
#Q23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_ = 5  # Average rate of occurrences (mean)
k_max = 15  # Maximum value to calculate CDF for

# Calculate the CDF for k values from 0 to k_max
k_values = np.arange(0, k_max+1)
cdf_values = poisson.cdf(k_values, lambda_)

# Plotting the CDF
plt.figure(figsize=(8, 6))
plt.step(k_values, cdf_values, where='post', label='Poisson CDF', color='blue', alpha=0.7)
plt.xlabel('Number of Events (k)', fontsize=14)
plt.ylabel('CDF', fontsize=14)
plt.title(f'Poisson CDF (Œª = {lambda_})', fontsize=16)
plt.grid(True)
plt.legend()

# Display the plot
plt.show()

# Printing CDF values for interpretation
for k, cdf in zip(k_values, cdf_values):
    print(f'P(X ‚â§ {k}) = {cdf:.4f}')
```

```python
#Q24. Simulate a random variable using a continuous uniform distribution and calculate its expected value.

import numpy as np

# Parameters for the continuous uniform distribution
a = 0  # Minimum value
b = 10  # Maximum value
n = 10000  # Number of random samples to simulate

# Step 1: Simulate random samples from a continuous uniform distribution
random_samples = np.random.uniform(a, b, n)

# Step 2: Calculate the sample mean (which is the estimate of the expected value)
sample_mean = np.mean(random_samples)

# Step 3: Calculate the theoretical expected value
theoretical_expected_value = (a + b) / 2

# Display results
print(f"Sample Mean (Estimated Expected Value): {sample_mean:.4f}")
print(f"Theoretical Expected Value: {theoretical_expected_value:.4f}")
```

```python
#Q27. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

# Example dataset (can be any numeric data)
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Normally distributed data with mean=50, std=15

# Calculate mean and standard deviation of the dataset
mean = np.mean(data)
std_dev = np.std(data)

# Z-score normalization
z_scores = (data - mean) / std_dev

# Plot the original and normalized data distributions
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot original data distribution
ax1.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
ax1.set_title('Original Data Distribution')
ax1.set_xlabel('Value')
ax1.set_ylabel('Frequency')

# Plot normalized data distribution
ax2.hist(z_scores, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
ax2.set_title('Normalized Data (Z-score) Distribution')
ax2.set_xlabel('Z-score')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#Q27. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

# Example dataset (can be any numeric data)
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Normally distributed data with mean=50, std=15

# Calculate the mean and standard deviation of the dataset
mean = np.mean(data)
std_dev = np.std(data)

# Z-score normalization: (X - mean) / std_dev
z_scores = (data - mean) / std_dev

# Plot the original and normalized data distributions
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot original data distribution
ax1.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
ax1.set_title('Original Data Distribution')
ax1.set_xlabel('Value')
ax1.set_ylabel('Frequency')

# Plot normalized data (Z-scores) distribution
ax2.hist(z_scores, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
ax2.set_title('Normalized Data (Z-score) Distribution')
ax2.set_xlabel('Z-score')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#Q28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

# Function to calculate skewness and kurtosis
def calculate_skewness_kurtosis(data):
    # Calculate skewness
    skew_value = skew(data)

    # Calculate kurtosis (Fisher's definition, which subtracts 3 for normal distribution)
    kurtosis_value = kurtosis(data)

    return skew_value, kurtosis_value

# Parameters for the normal distribution
mean = 50  # Mean of the distribution
std_dev = 10  # Standard deviation of the distribution
size = 1000  # Size of the dataset

# Generate a dataset from a normal distribution
data = np.random.normal(mean, std_dev, size)

# Calculate skewness and kurtosis
skew_value, kurtosis_value = calculate_skewness_kurtosis(data)

# Output the results
print(f"Skewness: {skew_value:.4f}")
print(f"Kurtosis: {kurtosis_value:.4f}")
```
